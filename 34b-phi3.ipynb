# Companion notebook for finetune of Phi 3.5 with 3.5 billion parameters

Evaluation of the finetune with specific prompts about LLMs and GRPO.

```python
import torch
from unsloth import FastLanguageModel
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
model_name = 'microsoft/Phi-3.5-mini-instruct'
dataset_name = 'llm-abstract-dataset.jsonl.xz'
output_dir = "runs/" + model_name.split("/")[-1] + "-" + dataset_name.split(".")[0]
```

```python
model, tokenizer = FastLanguageModel.from_pretrained(model_name, 
    max_seq_length = 2048,
    dtype = None, # autodetect
    load_in_4bit = False)
```

```python
FastLanguageModel.for_inference(model)
```

```python
print(tokenizer.chat_template)
```

```python
messages = [{ "content": """You are an educated researcher and always answer in correct scientific terms.
                             You are very deep into AI and its methodologies. You are very creative.""", 
              "role": "system" }, 
            { "content": "Write an abstract with the title 'Frontiers in GRPO Finetuning'",
              "role": "user" },
            ]
```

```python
tokens = tokenizer.apply_chat_template(messages, add_generation_prompt=True,
                                       return_tensors = "pt").to("cuda")
```

```python
print(tokenizer.batch_decode(tokens)[0])
```

```python
outputs = model.generate(tokens, max_new_tokens = 512, use_cache = True,
                         do_sample=True, temperature=0.7, top_k=25, top_p=0.8)
print(tokenizer.batch_decode(outputs)[0])
```

```python
outputs = model.generate(tokens, max_new_tokens = 512, use_cache = True,
                         do_sample=True, temperature=0.7, top_k=25, top_p=0.8)
print(tokenizer.batch_decode(outputs)[0])
```

```python
from peft import PeftModel
model = PeftModel.from_pretrained(model, output_dir)
```

```python
model = model.merge_and_unload()
```
